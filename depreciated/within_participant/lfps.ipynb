```python
import numpy as np
import os
import globals as gl
import pandas as pd
import matplotlib.pyplot as plt
import mat73

plt.style.use('default')
```

```python
monkey = 'Pert'
path = os.path.join(gl.baseDir, 'smp2', 'LFPs', monkey)
lfp = np.load(os.path.join(path, 'lfp_aligned.PMd.npy'))
trial_info = pd.read_csv(os.path.join(path, 'trial_info.tsv'), sep='\t')
cfg = mat73.loadmat(path + '/cfg.mat')['cfg']
```

```python
fig, ax = plt.subplots()

tAx = np.linspace(0, lfp.shape[0], lfp.shape[0])

ax.pcolormesh(tAx, cfg['foi'], lfp.mean(axis=(2, 3)).T, shading='auto')
ax.set_yscale('log')
ax.set_xticks([20, 114])
ax.set_xticklabels(['Cue', 'Pert'])
ax.set_ylabel('frequency (Hz)')

ax.set_title('Average LFPs')

ax.axvline(20, color='w', lw=.8)
ax.axvline(84, color='w', lw=.8, ls=':')
ax.axvline(114, color='w', lw=.8)
ax.axhline(3, color='w', lw=.8, ls='--')
ax.axhline(8, color='w', lw=.8, ls='--')
ax.axhline(13, color='w', lw=.8, ls='--')
ax.axhline(25, color='w', lw=.8, ls='--')
ax.axhline(100, color='w', lw=.8, ls='--')

plt.show()
```
