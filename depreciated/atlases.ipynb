```python
import os
import matplotlib.pyplot as plt
import surfAnalysisPy as surf
import globals as gl
import nibabel as nb

from Plot import save_figure_incremental
```

```python
experiment = 'smp2'

atlas = 'ROI'

Hem = ['L', 'R']
hemispheres = ['Left hemisphere', 'Right hemisphere']
surface = ['fs32k_L', 'fs32k_R']

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

for h, H in enumerate(Hem):
    A = nb.load(os.path.join(gl.atlas_dir, f'{atlas}.32k.{Hem[h]}.label.gii'))
    plt.sca(axs[h])
    surf.plot.plotmap(A.darrays[0].data, surface[h],
                      underlay=None,
                      cmap='tab20',
                      borders=gl.borders[H],
                      underscale=[-1.5, 1],
                      alpha=.8,
                      new_figure=False,
                      colorbar=False)
    axs[h].set_title(hemispheres[h], fontsize=35)

fig.subplots_adjust(left=0, right=1, bottom=0, top=0.95, wspace=0, hspace=0)

save_figure_incremental(fig, f'atlas_{atlas}')
```
