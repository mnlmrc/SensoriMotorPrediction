```python
import PcmPy as pcm
import numpy as np
import matplotlib.pyplot as plt
import globals as gl
import os
import warnings
from vis import pcm_spike, plot_D_lfp

warnings.filterwarnings("ignore")

plt.style.use('default')
```

```python
path = os.path.join(gl.baseDir, 'smp2', 'spikes', gl.pcmDir)
monkey = 'Malfoy'
cuePre = 0
cueIdx = 20
cuePost = 84
pertIdx = 114
pertPost = 134

recordings = {
    'PFC': [17, 18, 19, 20, 21, 22, 23, 24],
    'PMd': [17, 18, 19, 20, 21, 22, 23, 24, 29], #
    'M1': [25, 27, 28],
    'S1': [26, 27, 28],}
```

```python
roi = 'PFC'
epoch = 'plan'
rec = recordings[roi]

fig, axs = plt.subplots(1, 2, sharex='col', sharey=True, figsize=(3, 2.5))

vmin, vmax = 0, 3

G_obs = []
for r in rec:
    G_obs.append(np.load(os.path.join(path, f'G_obs.spike.{monkey}.{roi}.aligned.{epoch}-{r}.npy')))
G_obs = np.array(G_obs).mean(axis=0)
ax, h = plot_D_lfp(fig, axs, 0, G_obs[:cueIdx].mean(axis=0), list(gl.regressor_mapping.keys())[:5], vmin=vmin, vmax=vmax, sqrt=True)
ax, _ = plot_D_lfp(fig, axs, 1, G_obs[cueIdx:cuePost].mean(axis=0), list(gl.regressor_mapping.keys())[:5], vmin=vmin, vmax=vmax, sqrt=True)
cbar = fig.colorbar(h, ax=axs, fraction=0.02, pad=0.04)
cbar.set_label('dissimilarity (a.u.)')
axs[0].set_title('Pre')
axs[1].set_title('Post')

fig.suptitle(f'Dissimilarities between\nspiking patterns (monkey {monkey[0]}, {roi})')

fig, axs = plt.subplots(1, 2, sharex='col', sharey=True, figsize=(5, 3), gridspec_kw={'width_ratios': [3, .5]})
_, axs = pcm_spike(fig, axs, roi, epoch, monkey, rec)
axs[0].set_ylim(-.1, 3)
axs[0].spines['left'].set_bounds(0, 3)

plt.show()
```
