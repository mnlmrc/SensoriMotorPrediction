```python
import globals as gl
import pickle
import os
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')
```

```python
baseDir = '/cifs/pruszynski/Marco/SensoriMotorPrediction'
lfpDir = 'LFPs' #+ monkey + '/'
spkDir = 'spikes' #+ monkey + '/'
pcmDir = 'pcm'
```

```python
from vis import plot_pcm_corr

f = open(os.path.join(gl.baseDir, 'smp2', gl.pcmDir, f'M.plan-exec.p'), "rb")
Mflex = pickle.load(f)
dtype = ['lfp', 'spk']
title = ['LFPs', 'Spiking activity']
fig, axs = plt.subplots(1, len(dtype), sharex=True, sharey=True, constrained_layout=True, figsize=(3, 4))

freq = 'alpha'

for d, dt in enumerate(dtype):
    print(d)
    fname = f'{dt}.corr_cue-dir.M1-S1.{freq}' if dt=='lfp' else f'{dt}.corr_cue-dir.M1-S1'
    f = open(os.path.join(baseDir, pcmDir, f'theta_in.{fname}.p'), 'rb')
    theta = pickle.load(f)[0]
    r_bootstrap = np.load(os.path.join(baseDir, pcmDir, f'r_bootstrap.{fname}.npy'))
    f = open(os.path.join(baseDir, pcmDir, f'theta_gr.{fname}.p'), 'rb')
    theta_g = pickle.load(f)[0]
    fig, axs = plot_pcm_corr(fig, axs, d, Mflex, theta, theta_g, r_bootstrap)
    axs[d].set_title(title[d])

fig.supxlabel('SNR')
axs[0].set_ylabel('correlation')

fig.suptitle(f'Cue-direction correlation in M1-S1 (alpha-band)')

plt.show()
```
