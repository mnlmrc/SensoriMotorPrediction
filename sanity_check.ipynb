```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import globals as gl
import os

plt.style.use('ggplot')
```

```python
experiment = 'smp2'
sn = 107

dat = pd.read_csv(os.path.join(gl.baseDir, experiment, gl.behavDir, f'subj{sn}', f'{experiment}_{sn}.dat'), sep='\t')

dat['startTR_diff'] = dat.startTRReal - dat.startTimeReal / 1000

fig, axs = plt.subplots(2, 1, sharex=True)
sns.lineplot(data=dat, ax=axs[0], x='TN', y='startTRReal', hue='BN', palette='Set2',)
sns.lineplot(data=dat, ax=axs[1], x='TN', y='startTR_diff', hue='BN', palette='Set2', legend=False)

fig.legend(*axs[0].get_legend_handles_labels(), loc='center left', bbox_to_anchor=(1, .5))
axs[0].legend_.remove()
fig.tight_layout()
```
